```python
# DecodeLabs AI Internship - Week 3 Project
# Done by: Sadaf Ramzan
# Content-Based Tech Stack Recommender System

import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("--- Week 3: Content-Based AI Recommendation Engine ---")

# Simulating our dataset
dataset = {
    'Job_Role': [
        'Data Scientist', 
        'DevOps Engineer', 
        'Backend Developer', 
        'Frontend Developer', 
        'Cloud Architect'
    ],
    'Required_Skills': [
        'python sql machine learning data analysis numpy pandas artificial intelligence',
        'aws docker kubernetes git ci/cd automation linux bash infrastructure',
        'java python sql apis databases nodejs express framework git backend architecture',
        'javascript react nextjs html css tailwind frontend UI web design typescript',
        'aws azure cloud computing infrastructure network security terraform virtualization'
    ]
}

df = pd.DataFrame(dataset)

# Programmatically creating the raw_skills.csv file locally so it's documented!
df.to_csv('raw_skills.csv', index=False)
print(" Career Tech-Stack dataset initialized and 'raw_skills.csv' generated locally.")

# Viewing the dataset inside notebook beautifully
df
```

```python
# Step 1 of IPO Pipeline: Taking 3 user inputs dynamically
print(" Welcome to the Digital Matchmaker! Please enter your top 3 current skills.")

user_inputs = []
for i in range(1, 4):
    skill = input(f"Enter Skill/Interest {i}: ").strip().lower()
    user_inputs.append(skill)

# Combining into a single vector-ready string
user_profile_str = " ".join(user_inputs)
print(f"\n Ingested User State Profile: '{user_profile_str}'")
```

```python
# Step 2: TF-IDF Vectorization
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['Required_Skills'])
user_vector = vectorizer.transform([user_profile_str])

# Step 3: Cosine Similarity Engine
similarity_scores = cosine_similarity(user_vector, tfidf_matrix).flatten()

# Step 4: Sorting & Filtering (Top-3 list to avoid choice overload)

df['Match_Percentage'] = similarity_scores * 100
ranked_results = df.sort_values(by='Match_Percentage', ascending=False)
top_3_recommendations = ranked_results.head(3)

print("="*55)
print(" TOP 3 RANKED CAREER PATH RECOMMENDATIONS")
print("="*55)

rank = 1
for index, row in top_3_recommendations.iterrows():
    print(f"{rank}.  Role: {row['Job_Role']}")
    print(f"    Angular Similarity Match: {row['Match_Percentage']:.2f}%")
    print(f"    Core Toolkit Framework: {row['Required_Skills']}\n")
    rank += 1

print("="*55)
```
